```python
mu, sigma = 3, 2
matrix = mu + sigma * np.random.randn(6, 6)
u, s, v = np.linalg.svd(matrix)
#matrix
```

```python
u @ np.diag(s) @ v
```

### task #2

```python
import imageio
import scipy.ndimage
import cv2
```

```python
image = imageio.imread("image.jpg")
background = imageio.imread("background.jpg")
```

```python
print("Image", image.shape)
print("Background", background.shape)
```

```python
back = cv2.resize(background, (1280, 720))
```

```python
lower_green = np.array([0, 180, 0])
upper_green = np.array([100, 255, 100])
```

```python
mask = cv2.inRange(image, lower_green, upper_green)
plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(image)
masked_image[mask != 0] = [0,0,0]
plt.imshow(masked_image)
```

```python
bgr_image = back.copy()
bgr_image[mask == 0] = [0,0,0]
plt.imshow(bgr_image)
```

```python
final_image = bgr_image + masked_image
plt.imshow(final_image)
```

### task #3

```python
import scipy.optimize

func = lambda x: x**4 + 9*x**3 + 16*x**2 - 36*x - 8
```

```python
# глобальный минимум функции f(x)
scipy.optimize.basinhopping(func, 0, T=1.0, niter=100)
```

```python
# корни уравнения f(x) = 0
scipy.optimize.root(func, [0, 5])
```

```python
x = sympy.symbols('x')
y = sympy.Function('y')
yt = y(x).diff(x)
eq = sympy.Eq(x**2 * yt + x * y(x) + x**2 * y(x)**2, 4)
eq
```

```python
solution = sympy.dsolve(eq, y(x))
solution
```
